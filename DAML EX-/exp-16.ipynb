```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from scipy.cluster.hierarchy import dendrogram, linkage

# Step 1: Load dataset
iris = pd.read_csv(r"C:\Users\7260k\Downloads\IRIS.csv")

# Step 2: Select only numeric columns (features)
x = iris.iloc[:, [0, 1, 2, 3]].values   # Sepal length, Sepal width, Petal length, Petal width

# Step 3: Apply KMeans clustering (3 clusters for Iris dataset)
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans.fit(x)

# Step 4: Scatter plot with clusters
plt.scatter(x[kmeans.labels_ == 0, 0], x[kmeans.labels_ == 0, 1], s=50, c='red', label='Setosa')
plt.scatter(x[kmeans.labels_ == 1, 0], x[kmeans.labels_ == 1, 1], s=50, c='green', label='Versicolor')
plt.scatter(x[kmeans.labels_ == 2, 0], x[kmeans.labels_ == 2, 1], s=50, c='blue', label='Virginica')

# Plot centroids
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=150, c='yellow', marker='X', label='Centroids')

plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.legend()
plt.show()

# Step 5: Dendrogram
dendrogram(linkage(x, method='ward'))
plt.title("Dendrogram for Iris Dataset")
plt.xlabel("Samples")
plt.ylabel("Euclidean Distance")
plt.show()
```
